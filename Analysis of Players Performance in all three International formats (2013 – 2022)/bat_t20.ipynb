```python
import pandas as pd
```

```python
df = pd.read_csv(r'E:\inno\Cricket EDA\batting(t20).csv')
df
```

```python
df[['Player', 'Team']] = df['player'].str.split('(', n=1, expand=True)
df
```

```python
df['Team'] = df['Team'].str.replace(')','')
df
```

```python
df = df.drop(columns=['player'])
```

```python
df
```

```python
df.columns
```

```python
df =  df.reindex(columns=['Player', 'Team', 'matches', 'inn', 'no', 'runs', 'hs', 'avg', 'bf', 'sr', '_100', '_50',
       '_0', '_4s', '_6s', 'season'])
```

```python
df
```

```python
df = df.rename(columns={'inn':'innings',
                   'no':'Not_out',
                   'hs':'Highest_score',
                   'avg':'average',
                   'bf':'Balls_faced',
                   'sr':'Strike_rate',
                   '_100':'100s',
                   '_50':'50s',
                   '_0':'Duck_out',
                   '_4s':'4s',
                   '_6s':'6s',
                   'season':'Year'})
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
col = ['Not_out','Balls_faced', '100s', '50s', 'Duck_out', '4s', '6s']
for i in col:
    df[i] = df[i].fillna(0)
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset = ['runs'])
```

```python
df.isna().sum()
```

```python
df['Dismissals'] = df['innings'] - df['Not_out']
df['average'] = round(df['runs']/df['Dismissals'],2)
df['Strike_rate'] = round((df['runs']/df['Balls_faced'])*100,2)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['Year'] = df['Year'].ffill()
```

```python
df.isna().sum()
```

```python
df
```

```python
df.info()
```

```python
df["Highest_score"] = df["Highest_score"].str.replace("*", "", regex=False).astype(int)
df["average"] = pd.to_numeric(df["average"], errors="coerce")
```

```python
col = ['Not_out', 'runs', 'Balls_faced', '100s',	'50s', 'Duck_out', '4s', '6s', 'Year', 'Dismissals']
for i in col:
    df[i] = df[i].astype('int')
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df["Boundary%"] = round(((df["4s"] * 4) + (df["6s"] * 6)) / df["runs"] * 100,2)
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['Team'].nunique()
```

```python
df['Team'].unique()
```

```python
plt.figure(figsize=(10, 6))
plt.suptitle("Univariate Analysis of T20 Batting Performance", fontsize=16)

plt.subplot(2, 2, 1)
sns.histplot(df["average"], bins=10, kde=True, color="darkblue")
plt.title("Distribution of Batting Average")
plt.xlabel("Batting Average")
plt.ylabel("Frequency")

plt.subplot(2, 2, 2)
sns.boxplot(x=df["Strike_rate"], color="lightblue")
plt.title("Boxplot of Strike Rate")
plt.xlabel("Strike Rate")

plt.subplot(2, 2, 3)
sns.histplot(df["runs"], bins=10, kde=True, color="darkgreen")
plt.title("Distribution of Runs")
plt.xlabel("Total Runs")
plt.ylabel("Frequency")

plt.subplot(2, 2, 4)
sns.kdeplot(df["Highest_score"], color="darkblue")
plt.title("Density Plot of Highest Individual Score")
plt.xlabel("Highest Score")

plt.tight_layout()
plt.show()
```

```python
top_10_scorers = df.nlargest(10, "runs")[["Player", "Team", "runs"]]

plt.figure(figsize=(5, 3))
sns.barplot(x=top_10_scorers["runs"], y= top_10_scorers["Player"], color= 'darkblue')
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.title("Top 10 Players by Runs in T20")
plt.show()

top_10_scorers
```

```python
total_runs_by_team = df.groupby('Team')[['runs']].sum().reset_index().sort_values(by= 'runs', ascending= False).head(10)
plt.figure(figsize=(7, 5))
sns.barplot(x= total_runs_by_team['runs'], y= total_runs_by_team['Team'], color= 'darkblue')
plt.xlabel("Total Runs")
plt.ylabel("Team")
plt.title("Total runs by Team in T20")
plt.show()
```

```python
import numpy as np
df = df.replace([np.inf, -np.inf], np.nan).dropna()
```

```python
yearly_stats = df.groupby(['Year'])[['average', 'Strike_rate']].mean().reset_index()

plt.figure(figsize=(8,3))
sns.lineplot(x="Year", y="average", data=yearly_stats, marker="o", label="Batting Average", color="darkblue")
sns.lineplot(x="Year", y="Strike_rate", data=yearly_stats, marker="s", label="Strike Rate", color="darkgreen")

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Yearly Trend of Batting Average and Strike Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_highest_scores = df.nlargest(10, "Highest_score")[["Player", "Team", "Highest_score"]]

plt.figure(figsize=(10, 5))
sns.barplot(x="Highest_score", y="Player", data=top_highest_scores, color= 'darkblue')
plt.xlabel("Highest Individual Score")
plt.ylabel("Player")
plt.title("Top 10 Highest Individual Scores in ODI")
plt.show()

top_highest_scores
```

```python
top_scores_yearly = df.loc[df.groupby("Year")["Highest_score"].idxmax(), ["Year", "Player", "Team", "Highest_score"]]

plt.figure(figsize=(7, 5))
top_player = top_scores_yearly.iloc[5]["Year"]
colors = ["darkgreen" if player == top_player else "darkblue" for player in top_scores_yearly["Year"]]
sns.barplot(x='Year', y="Highest_score", data=top_scores_yearly, palette= colors)

plt.xlabel("Year")
plt.ylabel("Highest Individual Score")
plt.title("Highest Individual Scores Per Year in T20")
plt.xticks(rotation=45)
plt.show()

top_scores_yearly
```

```python
yearly_boundaries = df.groupby('Year')['Boundary%'].mean().reset_index()

plt.figure(figsize= (7,3))
sns.lineplot(x="Year", y="Boundary%", data=yearly_boundaries, marker="o", color="darkgreen", label="Boundary %")

plt.xlabel("Year")
plt.ylabel("Boundary Percentage")
plt.title("Yearly Trend of Boundary Percentage in T20")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(3,3))
plt.pie(x=[df['6s'].sum(), df['4s'].sum()], labels=['6s', '4s'], colors=['darkblue', 'darkgreen'], autopct='%1.1f%%', explode=(0.1,0), shadow=True)
plt.title('contribution of Boundaries in T20')
plt.show()
```

```python
_100s = df.groupby('Player')[['100s']].sum().reset_index().sort_values(by= '100s', ascending=False)
print(_100s.head(10))
print('----------------------------------------------')
_50s = df.groupby('Player')[['50s']].sum().reset_index().sort_values(by= '50s', ascending=False)
print(_50s.head(10))
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
top_player = _100s.iloc[5]["Player"]
colors = ["darkgreen" if player == top_player else "darkblue" for player in _100s["Player"]]
sns.barplot(x=_100s['Player'][:5], y=_100s['100s'], palette= colors)
plt.title("Most number of centuries in T20")
plt.xlabel("Player_name")
plt.ylabel("No.of centuries")
plt.xticks(rotation=45)

plt.subplot(2,2,2)
top_player = _50s.iloc[1]["Player"]
colors = ["darkgreen" if player == top_player else "darkblue" for player in _50s["Player"]]
sns.barplot(x=_50s['Player'][:5], y= _50s['50s'], palette= colors)
plt.title("Most number of Half-centuries in T20")
plt.xlabel("Player_name")
plt.ylabel("No.of centuries")
plt.xticks(rotation=45)
```

```python
top_scorers = df.groupby(["Year", "Player"]).agg({
    "runs": "sum",
    "average": "mean",
    "Strike_rate": "mean",
    "Highest_score": "max",
    "100s": "sum",
    "50s": "sum"
}).reset_index()

top_scorers_by_season = top_scorers.loc[top_scorers.groupby("Year")["runs"].idxmax()]
top_scorers_by_season
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))
sns.barplot(x= top_scorers_by_season['Player'], y= top_scorers_by_season['runs'], color='darkblue', ax=ax1)
ax1.set_ylabel("Total Runs")
ax1.set_xlabel("Player")
plt.xticks(rotation=45)


ax2 = ax1.twinx()
sns.lineplot(x=top_scorers_by_season['Player'], y=top_scorers_by_season['average'], marker='s', color = 'darkred', label='Average', ax=ax2)
sns.lineplot(x=top_scorers_by_season['Player'], y=top_scorers_by_season['Strike_rate'], marker= 'o', color = 'darkgreen', label='Strike_rate', ax=ax2)
ax2.set_ylabel("Average and Strike rate")
ax2.set_xlabel("Player")
plt.xticks(rotation=45)
```

```python
df["Weighted_Score"] = (df["runs"] * 0.5) + (df["Strike_rate"] * 0.3) + (df["average"] * 0.2)
top_players = df.nlargest(10, "Weighted_Score")[["Player", "Team", "Weighted_Score"]]
top_players
```
