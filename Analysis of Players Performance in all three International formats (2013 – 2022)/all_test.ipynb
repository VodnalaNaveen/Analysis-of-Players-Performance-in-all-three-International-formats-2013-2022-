```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# Runs Distribution
plt.subplot(2, 2, 1)
sns.histplot(df["runs"], bins=20, kde=True, color="blue")
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.ylabel("Frequency")

# Average Distribution
plt.subplot(2, 2, 2)
sns.histplot(df["batting_avg"], bins=20, kde=True, color="red")
plt.title("Distribution of Batting Average")
plt.xlabel("Average")
plt.ylabel("Frequency")

# Strike Rate Distribution
plt.subplot(2, 2, 3)
sns.histplot(df["bowling_avg"], bins=20, kde=True, color="green")
plt.title("Distribution of bowling average")
plt.xlabel("Strike Rate")
plt.ylabel("Frequency")

# Wickets Distribution
plt.subplot(2, 2, 4)
sns.histplot(df["wickets"], bins=20, kde=True, color="purple")
plt.title("Distribution of Wickets")
plt.xlabel("Wickets")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
top_10_scorers = df.nlargest(10, "runs")[["player", "Team", "runs"]]
top_10_wickets = df.nlargest(10, 'wickets')[['player', 'Team', 'wickets']]

plt.figure(figsize=(5, 3))
sns.barplot(x="runs", y="player", data=top_10_scorers, color= 'darkblue')
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.title("Top 10 Players by Runs in Test")
plt.show()

plt.figure(figsize=(5, 3))
sns.barplot(x="wickets", y="player", data=top_10_wickets, color= 'darkblue')
plt.xlabel("Total Runs")
plt.ylabel("Player")
plt.title("Top 10 Players by wickets in Test")
plt.show()

print(top_10_scorers)
print('----------------------------------------------')
print(top_10_wickets)
```

```python
df['Team']= df['Team'].str.replace('1 3 (','')
```

```python
df['Team']= df['Team'].str.replace('3 (','')
```

```python
total_runs_by_team = df.groupby('Team')[['runs']].sum().reset_index()
total_runs_by_team
plt.figure(figsize=(7, 5))
sns.barplot(x=total_runs_by_team['runs'], y= total_runs_by_team['Team'], color= 'darkblue')
plt.xlabel("Total Runs")
plt.ylabel("Team")
plt.title("Total runs by Team in Test")
plt.show()
```

```python
yearly_stats = df.groupby(['year'])[['batting_avg', 'bowling_avg']].mean().reset_index()

plt.figure(figsize=(8,3))
sns.lineplot(x="year", y="batting_avg", data=yearly_stats, marker="o", label="Batting Average", color="darkblue")
sns.lineplot(x="year", y="bowling_avg", data=yearly_stats, marker="s", label="Strike Rate", color="darkgreen")

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("Yearly Trend of Batting Average and Strike Rate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
top_scorers = df.loc[df.groupby('year')['runs'].idxmax()][['year', 'player', 'runs']]
top_wicket_takers = df.loc[df.groupby('year')['wickets'].idxmax()][['year', 'player', 'wickets']]

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel("Year (Season)")
ax1.set_ylabel("Top Runs", color="blue")
ax1.plot(top_scorers['year'], top_scorers['runs'], color="blue", marker="o", label="Top Run-Scorer")

for i, txt in enumerate(top_scorers['player']):
    ax1.annotate(txt, 
                 (top_scorers['year'].iloc[i], top_scorers['runs'].iloc[i]), 
                 fontsize=9, color="darkblue", 
                 xytext=(0, 8), textcoords="offset points", ha="center")

ax2 = ax1.twinx()
ax2.set_ylabel("Top Wickets", color="red")
ax2.plot(top_wicket_takers['year'], top_wicket_takers['wickets'], color="red", marker="s", label="Top Wicket-Taker")

for i, txt in enumerate(top_wicket_takers['player']):
    ax2.annotate(txt, 
                 (top_wicket_takers['year'].iloc[i], top_wicket_takers['wickets'].iloc[i]), 
                 fontsize=9, color="red", 
                 xytext=(0, -12), textcoords="offset points", ha="center")

fig.suptitle("Top Run-Scorer & Wicket-Taker Each Year (Test)")
fig.legend(loc="upper left")

plt.xticks(rotation=45)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
top_scores_yearly = df.loc[df.groupby("year")["avg_diff"].idxmax(), ["year", "player", "Team", 'runs', 'batting_avg', 'wickets', 'bowling_avg', 'best_bowling_innings', "avg_diff"]]

fig, ax1 = plt.subplots(figsize=(8, 4))
top_player = top_scores_yearly.iloc[4]["year"]
colors = ["darkgreen" if player == top_player else "darkblue" for player in top_scores_yearly["year"]]
sns.barplot(x='player', y="avg_diff", data=top_scores_yearly, palette= colors)
ax1.set_ylabel("avg_difference")
ax1.set_xlabel("Player")
plt.xticks(rotation=45)

ax2 = ax1.twinx()
sns.lineplot(x= top_scores_yearly['player'], y= top_scores_yearly['batting_avg'], marker='s' ,color = 'darkred', label='bating_average', ax=ax2)
sns.lineplot(x= top_scores_yearly['player'], y= top_scores_yearly['bowling_avg'], marker='o' ,color = 'green', label='bowling_average', ax=ax2)
ax2.set_ylabel("Bowling and Batting Average")
ax2.set_xlabel("Player")

plt.title("average difference Per player in ODI")
plt.xticks(rotation=45)
plt.show()

top_scores_yearly
```
