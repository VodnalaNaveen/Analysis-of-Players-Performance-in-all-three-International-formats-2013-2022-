```python
import pandas as pd
```

```python
df = pd.read_csv(r'E:\inno\Cricket EDA\batting(odi).csv')
df
```

```python
df[['Player', 'Team']] = df['player'].str.split('(', n=1, expand=True)
df
```

```python
df['Team'] = df['Team'].str.replace(')','')
df
```

```python
df = df.reindex(columns = ['Player', 'Team', 'player', 'matches', 'inn', 'no', 'runs', 'hs', 'avg', 'bf', 'sr', '_100', '_50', '_0', '_4s', '_6s', 'season'] )
df
```

```python
df = df.drop(columns= 'player')
df
```

```python
df = df.rename(columns={'inn':'innings',
                   'no':'Not_out',
                   'hs':'Highest_score',
                   'bf':'Balls_faced',
                   'avg':'average',
                   'sr':'Strike_rate',
                   '_100':'100s',
                   '_50':'50s',
                   '_0':'Duck_out',
                   '_4s':'4s',
                   '_6s':'6s',
                   'season':'Year'})

df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
col = ['Not_out','Balls_faced', '100s', '50s', 'Duck_out', '4s', '6s']
for i in col:
    df[i] = df[i].fillna(0)
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset = ['runs'])
```

```python
df.isna().sum()
```

```python
df['Dismissals'] = df['innings'] - df['Not_out']
df['average'] = round(df['runs']/df['Dismissals'],2)
df['Strike_rate'] = round((df['runs']/df['Balls_faced'])*100,2)
```

```python
df.isna().sum()
```

```python
df['Year'] = df['Year'].ffill()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df["Highest_score"] = df["Highest_score"].str.replace("*", "", regex=False).astype(int)
df["average"] = pd.to_numeric(df["average"], errors="coerce")
```

```python
col = ['Not_out', 'runs', 'Balls_faced', '100s', '50s', 'Duck_out', '4s', '6s', 'Year', 'Dismissals']
for i in col:
    df[i] = df[i].astype('int')
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df
```

```python
df["Boundary%"] = round(((df["4s"] * 4) + (df["6s"] * 6)) / df["runs"] * 100,2)
```

```python

top_scorers = df.groupby(["Year", "Player"]).agg({
    "runs": "sum",
    "average": "mean",
    "Strike_rate": "mean",
    "Highest_score": "max",
    "100s": "sum",
    "50s": "sum"
}).reset_index()

top_scorers_by_season = top_scorers.loc[top_scorers.groupby("Year")["runs"].idxmax()]
top_scorers_by_season
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4))
sns.barplot(x= top_scorers_by_season['Player'], y= top_scorers_by_season['runs'], color= 'darkblue', ax=ax1)
ax1.set_ylabel("Total Runs")
ax1.set_xlabel("Player")
plt.xticks(rotation=45)


ax2 = ax1.twinx()
sns.lineplot(x=top_scorers_by_season['Player'], y=top_scorers_by_season['average'], marker='s' ,color = 'darkred', label='Average', ax=ax2)
sns.lineplot(x=top_scorers_by_season['Player'], y=top_scorers_by_season['Strike_rate'], marker='o' ,color = 'darkgreen', label='Strike_rate', ax=ax2)
ax2.set_ylabel("Average and Strike rate")
ax2.set_xlabel("Player")

ax2.legend()
plt.show()
```

* Weight Breakdown:
* 40% - Batting Average (Most crucial metric)
* 30% - Strike Rate (Reflects scoring speed)
* 20% - Runs per Innings (Consistency)
* 10% - Boundary Percentage (Aggression)
