```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_t20=pd.read_csv(r'C:\Users\vodna\OneDrive\Desktop\p\bowling(T20).csv')
df_t20
```

```python
df_t20.isna().sum()
```

```python
df_t20=df_t20.dropna()
```

```python
df_t20.info()
```

```python
df_t20=df_t20.rename(columns={'inn':'innings',
                       'wkts':'wickets',
                       'bb_inn':'best_bowling_innings',
                       'avg':'average',
                       'econ':'economy',
                       'sr':'strike rate',
                       "_5w": "5wicket_haul",
                       "_4w": "4wicket_haul",
                       'runs':'runs conceded',
                       'season':'year'})
df_t20
```

```python
df_t20.columns
```

```python
column=['maiden', 'runs conceded', 'wickets','4wicket_haul', '5wicket_haul', 'year']
for i in column:
    df_t20[i]=df_t20[i].astype('int')
```

```python
df_t20[['player','country']]=df_t20['player'].str.split('(',n=1,expand=True)
df_t20
```

```python
df_t20['country']=df_t20['country'].str.replace(')','')
```

```python
df_t20["balls_bowled"] = df_t20["overs"].astype(str).str.split(".").apply(lambda x: int(x[0]) * 6 + int(x[1]) if len(x) > 1 else int(x[0]) * 6)
df_t20
```

```python
df_t20=df_t20[['player', 'country', 'matches', 'innings', 'overs','balls_bowled', 'maiden', 'runs conceded', 'wickets',
       'best_bowling_innings', 'average', 'economy', 'strike rate',
       '4wicket_haul', '5wicket_haul', 'year']]

df_t20
```

```python
df_t20['weight score']=(0.4*df_t20['wickets'])-(0.3*df_t20['average'])-(0.2*df_t20['economy'])-(0.1*df_t20['strike rate'])
df_t20
```

```python
fig,ax=plt.subplots(2,2,figsize=(15, 7))
plt.suptitle("Univariate Analysis of ODI Bowling Performance", fontsize=16)

plt.subplot(2, 2, 1)
sns.histplot(df_t20['average'],bins=10,ax=ax[0,0],kde=True,color='darkblue')
plt.title('Distribution of Bowling Average')
plt.xlabel('Average')
plt.ylabel('Values')

plt.subplot(2, 2, 2)
sns.boxplot(x=df_t20['strike rate'],ax=ax[0,1],color='lightblue')
plt.title('Boxplot of Strike Rate')
plt.xlabel('Strike Rate')

plt.subplot(2, 2, 3)
sns.histplot(df_t20['runs conceded'],bins=10,ax=ax[1,0],kde=True,color='darkgreen')
plt.title('Distribution of Runs Conceded')
plt.xlabel('Runs Conceded')
plt.ylabel('Values')

plt.subplot(2, 2, 4)
sns.boxplot(x=df_t20['economy'],ax=ax[1,1],color='lightblue')
plt.title('Boxplot of Economy Rate')
plt.xlabel('Economy Rate')

plt.tight_layout()
plt.show()
```

```python
top_wicket_taker_player=df_t20.groupby('player')[['wickets']].max().reset_index().sort_values(by='wickets',ascending=False).head(10)

plt.figure(figsize=(12,4))

sns.barplot(y=top_wicket_taker_player['player'],x=top_wicket_taker_player['wickets'],color='darkblue')
plt.title("Top 10 wicket takers in ODI")
plt.show()
top_wicket_taker_player
```

```python
top_wicket_by_country=df_t20.groupby('country')[['wickets']].max().reset_index().sort_values(by='wickets',ascending=False).head(10)
plt.figure(figsize=(10,4))

sns.barplot(data=top_wicket_by_country,x='wickets',y='country',color='darkblue')
plt.title('Top 10 wickets by country wise')
plt.show()
top_wicket_by_country
```

```python
avg_eco_sr_trend=df_t20.groupby('year')[['average','economy','strike rate']].mean().reset_index()

plt.figure(figsize=(10,4))
sns.lineplot(data=avg_eco_sr_trend,x='year',y='average',marker='o',label='average',color='darkblue')
sns.lineplot(data=avg_eco_sr_trend,x='year',y='economy',marker='s',label='economy',color='red')
sns.lineplot(data=avg_eco_sr_trend,x='year',y='strike rate',marker='^',label='strike rate',color='darkgreen')

plt.grid(True)
plt.xlabel('year')
plt.ylabel('values')
plt.title('Yearly Trend of Bowling Average,Economy and Strike Rate')
plt.legend()
plt.show()

avg_eco_sr_trend
```

```python
best_fig=df_t20.groupby('player')[['best_bowling_innings']].max().reset_index().sort_values(by='best_bowling_innings',ascending=False).head(10)

best_fig
```

```python
top_wicket_year=df_t20.groupby('year')['wickets'].max().reset_index()

plt.figure(figsize=(10,4))
top_player = top_wicket_year.iloc[9]["year"]
colors = ["darkgreen" if player == top_player else "darkblue" for player in top_wicket_year["year"]]
sns.barplot(data=top_wicket_year,x='year',y='wickets',palette=colors)
plt.title('Highest Wickets By Year')
plt.show()

top_wicket_year
```

```python
top_maiden=df_t20.groupby('player')['maiden'].max().reset_index().sort_values(by='maiden',ascending=False).head(10)

plt.figure(figsize=(12,4))
sns.barplot(data=top_maiden,x='player',y='maiden',color='darkblue')
plt.title('Top 10 Maiden overs bowled')
plt.show()
top_maiden
```

```python
df_t20[['4wicket_haul','5wicket_haul']].sum()

plt.pie(x=[df_t20['4wicket_haul'].sum(),df_t20['5wicket_haul'].sum()],labels=('4 wicket hauls','5 wicket hauls'),colors=['darkblue', 'darkgreen'], autopct='%1.1f%%', explode=(0.1,0), shadow=True)
plt.title('contribution of 4 & 5 wicket hauls in ODI')
```

```python
top_4wickets=df_t20.groupby('player')['4wicket_haul'].max().reset_index().sort_values(by='4wicket_haul',ascending=False).head()

plt.figure(figsize=(8,5))
sns.barplot(data=top_4wickets,x='player',y='4wicket_haul',color='darkblue')
plt.title('Top 5 4wicket hauls')
plt.show()

top_4wickets
```

```python
top_5wickets=df_t20.groupby('player')['5wicket_haul'].max().reset_index().sort_values(by='5wicket_haul',ascending=False).head()

plt.figure(figsize=(8,5))
sns.barplot(data=top_5wickets,x='player',y='5wicket_haul',color='darkblue')
plt.title('Top 5 5wicket hauls')
plt.show()

top_5wickets
```

```python
a=df_t20.groupby(['year','player','average','economy', 'strike rate'])['wickets'].max().reset_index()
b=a.loc[a.groupby('year')['wickets'].idxmax()]


fig, ax1 = plt.subplots(figsize=(15, 5))

sns.barplot(x=b['player'], y=b['wickets'], ax=ax1, color='darkblue', label='Wickets',errorbar=None)
ax1.set_ylabel("Wickets")
ax1.set_xlabel("Player")
plt.xticks(rotation=45)

ax2 = ax1.twinx()

sns.lineplot(x=b['player'], y=b['average'], marker='o', ax=ax2, label='Average', color='red',errorbar=None)
sns.lineplot(x=b['player'], y=b['economy'], marker='o', ax=ax2, label='Economy', color='green',errorbar=None)
sns.lineplot(x=b['player'], y=b['strike rate'], marker='o', ax=ax2, label='Strike Rate', color='purple',errorbar=None)

ax2.set_ylabel("Average,Economy and Strike rate", color='black')
ax2.set_xlabel("Player")
plt.xticks(rotation=45)

fig.suptitle("Year wise Top Bowling Performance")

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')






plt.show()

b
```

```python
best_10=df_t20.nlargest(10,'weight score')

plt.figure(figsize=(13,5))
sns.barplot(data=best_10,x='weight score',y='player',color='darkblue')
plt.title('Best 10 Players')
plt.show()

best_10
```
